## 경매 + 생활인구

### 생활인구 데이터 정리

```python
### 생활인구정리
population = pd.read_csv('4_생활인구.csv', engine='python', header='infer', encoding='UTF-8')
population = population.drop(['기준일ID', '시간대구분'], axis=1)
population_gruop = population.groupby("행정동코드").mean()
population_gruop.reset_index()
```

```python
# 법정동-행정동 코드
dong_mapping = pd.read_csv('3_code.csv', engine='python', header='infer', encoding='euc-kr')
dong_mapping['행정동코드'] = dong_mapping['행정동코드']//100
```

```python
# 생활인구 + 법정동행정동 merge
population_1 = pd.merge(dong_mapping, population_gruop, how = 'left', on = '행정동코드')
```

```python
population_1 = population_1.drop(['행정동코드'], axis=1)

# 생활인구, 법정동코드 남기기
```

```python
population_1.법정동코드 = population_1.법정동코드.astype('str')
```

```python
population_gruop2 = population_1.groupby('법정동코드').mean()
population_gruop2.reset_index()
```

### 경매(우편) + 생활인구 merge

```python
auc_pop = pd.merge(auc_post, population_gruop2, how = 'left', on = '법정동코드')
```

```python
# 결과 : 경매데이터 (+ 도로명, 법정데이터) + 생활인구
```

## 경매 + 유동인구

### 유동인구 데이터 정리

```python
### 유동인구 
flow_pop = pd.read_csv('2_building_code.csv', engine='python', header='infer', encoding = 'euc-kr')
flow_pop.columns = ['기준년월', '상권코드', '도로명', '유동인구']
flow_pop = flow_pop.drop(['기준년월', '상권코드'], axis = 1)

# 도로명, 유동인구만 남기기
```

```python
flow_pop.columns = ['도로명주소만', '유동인구']
```

```python
# 도로명주소인 유동인구데이터를 법정동주소 기준으로 변경
flow_pop = pd.merge(flow_pop, addr_code_2, how = 'left', on = '도로명주소만')
```

```python
flow_pop.columns
```

```python
flow_pop = flow_pop.drop(['도로명주소만', '우편번호'], axis = 1)
```

```python
flow_pop = flow_pop.groupby('법정동코드').mean()
flow_pop.reset_index()
```

### 경매(우편, 생활인구) + 유동인구 merge

```python
auc_flow = pd.merge(auc_pop, flow_pop, how = 'left', on = '법정동코드')
```

```python
# 결과 : 경매데이터 ((+ 도로명, 법정데이터) + 생활인구) + 유동인구
```

## 경매 + 매출

### 매출데이터 정리

```python
### 매출
sales = pd.read_csv('5_상권추정매출.csv', engine='python', header='infer', encoding = 'euc-kr')
sales.columns = ['도로명주소만', '매출금액', '매출건수']
```

```python
sales_group = sales.groupby(['도로명주소만']).mean()
print(sales_group.shape)
sales_group.reset_index().head()

# 도로명, 매출금액, 매출건수 남기기
```

```python
# 도로명주소인 매출데이터를 법정동주소 기준으로 변경
sales_group = pd.merge(sales_group, addr_code_2, how = 'left', on = '도로명주소만')
```

```python
sales_group.columns
```

```python
sales_group = sales_group.drop(['도로명주소만', '우편번호'], axis = 1)
```

```python
sales_group = sales_group.groupby('법정동코드').mean()
sales_group.reset_index()
```

```python
# 결과 : 경매데이터 (((+ 도로명, 법정데이터) + 생활인구) + 유동인구) + 매출
```
