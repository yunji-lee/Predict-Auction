## code 불러오기

```python
code = pd.read_csv('code.csv',engine='python', header = 'infer',encoding = 'euc-kr')
```

```python
code.head()
```

```python
code = code.drop(['생성일자', '말소일자'],axis=1)
```

```python
code.columns = ['hang_dond_code', 'si', 'gu', 'dong', 'bup_doing_code', 'bup_dong']
```

```python
code['city'] = code['si-do'] + " " + code['gu'] + " " + code['dong']
```

## code와 실거래 data 합치기

```python
pd.merge(df, code, how='left', on='city')
```
